```python
# creamos una clase llamada TiendaOnline
class TiendaOnline:
    def __init__ (self):
        self.inventario = [] #inventario (lista de diccionarios): el producto representado por un diccionario
        self.clientes = {} #clientes (diccionario)
        self.ventas_totales = 0 #ventas_totales (float)
        self.carrito = []

    # Método para agregar producto
    def agregar_producto(self, nombre, precio, cantidad):
        #Crear el diccionario del producto
        producto = {"nombre":nombre, "precio":precio, "cantidad":cantidad} 
        #Verifica si el producto ya existe en el inventario
        for item in self.inventario:
            if item ["nombre"] == nombre:
                #Actualizar la cantidad del producto existente
                item ["cantidad"] += cantidad
                print(f"El producto'{nombre}' ha sido actualizado. Nueva cantidad: 40")
                return
                print(f"El producto '{nombre}' ya existe")

        #Si el producto no existe, agregarlo al inventario
        self.inventario.append(producto)
        print(f"producto '{nombre}' agregado con éxito.")

    # Método para buscar un producto en el inventario (nombre)
    def buscar_producto(self, nombre):
        for item in self.inventario: 
            if item ["nombre"] == nombre: 
                return item
        return
    
    # Método actualizar stock
    def actualizar_stock (self, nombre, cantidad):
        for item in self.inventario:
            if item ["nombre"] == nombre:
                item["cantidad"]+= cantidad 
                print(f"el stock del producto '{nombre}' ha sido actualizado. Nueva cantidad: {item['cantidad']}")
                return
        print(f"No se encontro el producto '{nombre}'")

    # Método eliminar producto
    def eliminar_producto (self, nombre):
        for item in self.inventario:
            if item ["nombre"] == nombre: 
                del self.inventario[1]
                print(f"Producto'{nombre}'eliminado con éxito.")
                return
        print(f"no se encontro el producto '{nombre}' en el inventario")

    # Método calcular valor total inventario 
    def calcular_valor_inventario (self):
        valor_total = 0
        for producto in self.inventario:
           valor_total += producto["precio"] * producto["cantidad"]
        return valor_total


```

```python
#Crear una instancia de la clase TiendaOnline
tienda = TiendaOnline()

# Agregar productos al inventario
tienda.agregar_producto ("camisa", 20, 30)
tienda.agregar_producto ("pantalon", 30, 30)
tienda.agregar_producto ("camisa", 20, 10)
tienda.agregar_producto ("zapatos", 50, 40)
tienda.agregar_producto ("camiseta", 20, 50)

# Mostrar el inventario
print("Inventario:")
for producto in tienda.inventario:
    print(producto)


# Buscar un producto en el inventario 
producto_buscado = tienda.buscar_producto("camisa")
if producto_buscado:
    print(f"Producto encontrado: {producto_buscado}")
else:
    print("Producto no encontrado.")

#Actualizar el stock de un producto 
tienda.actualizar_stock ("camisa", 10)
tienda.actualizar_stock ("falda", 15)

# Eliminamos un producto del inventario
tienda.eliminar_producto("zapatos")

print ("inventario")
for producto in tienda.inventario:
    print(producto)

# calcular el valor total del inventario
valor_inventario = tienda.calcular_valor_inventario()
print(f"El valor total del inventario es ${valor_inventario}")

```
